```python
import os
im_dir = 'practice_exam_1_v2'
image_files = sorted(os.listdir(im_dir))
print(image_files)
```

```python
images = [plt.imread(im_dir+'/'+img_file) for img_file in image_files]
images = [img/np.amax(img) for img in images]
for img in images:
  plt.figure()
  plt.imshow(img)

inp_imgs = images[:-1]
out_img = images[-1]
```

```python
# Shows no balls
def q1(imgs):
  imgs = np.array(imgs) # Inp_imgs is a 4D array
  return np.median(imgs, axis = 0)

result = q1(inp_imgs) # Inp_imgs is a 4D array
plt.imshow(result)
```

```python
def q1(imgs):
  imgs = np.array(imgs) # Inp_imgs is a 4D array
  return np.max(imgs, axis = 0)

result = q1(inp_imgs) 
plt.imshow(result)
```

```python
def q1(imgs):
  imgs = np.array(imgs) # Inp_imgs is a 4D array
  return np.mean(imgs, axis = 0)

result = q1(inp_imgs) # Inp_imgs is a 4D array
plt.imshow(result)
```

```python
def q1(imgs):
  imgs = np.array(imgs) # Inp_imgs is a 4D array
  return np.max(imgs, axis = 0) * (1-(np.max(imgs, axis = 0) - np.min(imgs, axis = 0))) 

result = q1(inp_imgs) 
plt.imshow(result)
```

```python
def q1(imgs):
  imgs = np.array(imgs) # Inp_imgs is a 4D array
  return 1 - (np.max(imgs, axis = 0) - np.min(imgs, axis = 0))
result = q1(inp_imgs) 
plt.imshow(result)
```

```python
# Shows the 4 balls
def q2(imgs):
  imgs = np.array(imgs) # Inp_imgs is a 4D array
  return (np.sum(imgs, axis = 0) - 3*np.median(imgs,axis = 0))

result = q2(inp_imgs) 
plt.imshow(result)
```
